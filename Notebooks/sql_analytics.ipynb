```python
q1 = """
SELECT COUNT(*) AS customers,
       ROUND(AVG(Churn)*100, 2) AS churn_rate_pct,
       ROUND(AVG(MonthlyCharges), 2) AS avg_monthly_charges
FROM telco_churn;
"""
display(run_sql_query(DB_PATH, q1))
```

```python
q2 = """
SELECT Contract,
       COUNT(*) AS customers,
       ROUND(AVG(Churn)*100, 2) AS churn_rate_pct
FROM telco_churn
GROUP BY Contract
ORDER BY churn_rate_pct DESC;
"""
display(run_sql_query(DB_PATH, q2))
```
